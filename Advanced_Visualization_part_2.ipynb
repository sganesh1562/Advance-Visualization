```python
iris=snr.load_dataset("iris")
```

```python
df=iris.drop(['species'],axis=1)
```

```python
df.head()
```

```python
df.plot.scatter(x='sepal_length',y='petal_length')
```

```python
df.plot.scatter(x='sepal_width',y='petal_width')
```

```python
df.plot.scatter(x='sepal_width',y='petal_length')
```

```python
df.plot.scatter(x='sepal_length',y='petal_width')
```

```python
df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width')
```

```python
df.plot.hexbin(x='sepal_length',y='petal_length', gridsize=25,C='sepal_width')
```
